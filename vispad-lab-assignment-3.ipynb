### SECTION 1: ORTHOGONAL PROJECTION (5 Points)


Given the two 3D vectors, $\mathbf{v} = [1, 2, 3]$ and $\mathbf{y} = [2, 0, 2]$:


a. Compute the orthogonal projection of $\mathbf{v}$ onto $\mathbf{y}$ (i.e., $proj_{\mathbf{y}}(\mathbf{v})$).


b. Compute the orthogonal component $\mathbf{v}_{\perp}$ (the part of $\mathbf{v}$ that is perpendicular to $\mathbf{y}$).


c. Verify that your $\mathbf{v}_{\perp}$ from part (b) is truly orthogonal to $\mathbf{y}$ by computing their dot product.



```python
# Answer goes here
# Function to compute dot product (from previous lab)
def dot_product(v, y):
    total = 0
    for i in range(len(v)):
        total += v[i] * y[i]
    return total

# Function to compute orthogonal projection
def project_onto(v, y):
       # Calculate <v, y>
    numerator = dot_product(v, y)
    
    # Calculate <y, y>
    denominator = dot_product(y, y)
    
    # Calculate the scalar coefficient
    scalar = numerator / denominator
    
    # Multiply scalar by y
    projection = [scalar * y[i] for i in range(len(y))]
    
    return projection

v= [1,2,3]
y=[2,0,2]
print("the Projection of v onto y give:", project_onto(v, y))
```

```python
#b- the orthogonal component 
def orthogonal_component(v, y):
    proj = project_onto(v, y)
    
    # Subtract projection from original vector
    v_perp = [v[i] - proj[i] for i in range(len(v))]
    
    return v_perp
print("the orthogonal component is :", orthogonal_component(v, y))
```

```python
#c- Verification 
v_o= orthogonal_component(v, y)
if dot_product(v_o, y)== 0:
    print("Yes v_o is truly orthogonal to y")
else:
    print("No")
```

### SECTION 2: CALCULATING ANGLE (5 Points)

Given the two 3D vectors, $\mathbf{a} = [3, 4, 0]$ and $\mathbf{b} = [0, 0, -5]$:


a. Compute the angle (in degrees) between $\mathbf{a}$ and $\mathbf{b}$.


b. Now, compute the angle (in degrees) between $\mathbf{a} = [3, 4, 0]$ and $\mathbf{c} = [3, 4, 5]$.


```python
import numpy as np 
import math
# Answer goes here

def cosine_similarity(a, b):
    if len(a)!=len(b):
        raise ValueError("Vectors must have the same dimension")
    numerator= dot_product(a, b)
    norm_a = np.linalg.norm(a)
    norm_b= np.linalg.norm(b)
    denominator= norm_a * norm_b
    resutl= numerator / denominator
    return resutl


def angle_between_rad(a, b):
    
    cos_sim = cosine_similarity(a, b)
    cos_sim = min(max(cos_sim, -1.0), 1.0)
    
    # a. Calculate the angle in radians
    angle_rad = math.acos(cos_sim)
       
    return angle_rad

a = [3, 4, 0]
b = [0, 0, -5]

print ("the angle between a and b in radian is", angle_between_rad(a, b))

# b Calculate the angle in degree of a and b 
a= [3, 4, 0]
b= [3, 4, 5]
def angle_between_deg(a, b):
    # b. Convert to degrees
    cos_sim= cosine_similarity(a, b)
    cos_sim= min(max(cos_sim,-1.0),1.0)
    angle_rad= math.acos(cos_sim)
    angle_deg = math.degrees(angle_rad)
    return angle_deg

print ("the angle between a and b in degree is", angle_between_deg(a, b))
```

### SECTION 3: GRAM-SCHMIDT PROCESS (5 Points)

Given these vectors:

$\mathbf{v_1} = [2, 0, 0, 0]$

$\mathbf{v_2} = [1, 3, 0, 0]$

$\mathbf{v_3} = [1, 1, 4, 0]$

$\mathbf{v_4} = [2, 2, 1, 5]$

a. Apply the Gram-Schmidt process to find the orthogonal basis ${\mathbf{e_1}, \mathbf{e_2}, \mathbf{e_3}, \mathbf{e_4}}$. Print each of these new vectors.

b. Create the orthonormal basis ${\mathbf{u_1}, \mathbf{u_2}, \mathbf{u_3}, \mathbf{u_4}}$ by normalizing your orthogonal vectors from part (a). Print each of these unit vectors.


```python
# Answer goes here 

v_1= [2,0,0,0]
v_2= [1,3,0,0]
v_3= [1,1,4,0]
v_4= [2,2,1,5]
#a. Orthogonal basis e1,e2, e3, e4
def gram_schmidt(vectors):
  # Start with empty list for orthogonal vectors
    orthogonal = []
    
    for v in vectors:
        # Start with the current vector
        e = v.copy()
        
        # Subtract projections onto all previous orthogonal vectors
        for orth_vec in orthogonal:
            # Calculate projection
            proj = project_onto(e, orth_vec)
            
            # Subtract projection from current vector
            e = [e[i] - proj[i] for i in range(len(e))]
        
        # Add the new orthogonal vector to our list
        orthogonal.append(e)
    
    return orthogonal

vectors = [v_1, v_2, v_3, v_4]
orthogonal = gram_schmidt(vectors)

print("Original vectors:")
for i, v in enumerate(vectors):
    print(f"  v{i+1} = {v}")

print("\nOrthogonal vectors:")
for i, e in enumerate(orthogonal):
    print(f"  e{i+1} = {e}")

# Verify orthogonality (dot product should be ~0)
print(f"\nDot product e1Â·e2 = {dot_product(orthogonal[0], orthogonal[1]):.6f}")


""" b- Create the orthonormal basis u1, u2, u3, u4 """
v_1= [2,0,0,0]
v_2= [1,3,0,0]
v_3= [1,1,4,0]
v_4= [2,2,1,5]
vectors = [v_1, v_2, v_3, v_4]
def normalize(v):
    """Convert vector to unit length"""
    import math
    length = math.sqrt(sum(x**2 for x in v))
    return [x / length for x in v]

def gram_schmidt_orthonormal(vectors):
    """
    Apply Gram-Schmidt and normalize to create orthonormal basis
    """
    # First, get orthogonal vectors
    orthogonal = gram_schmidt(vectors)
    
    # Then normalize each one
    orthonormal = [normalize(e) for e in orthogonal]
    
    return(orthonormal)

orthonormal = gram_schmidt_orthonormal(vectors)
print("\nthe orthonormal basis:")
for i, u in enumerate(orthonormal):
    print(f" u{i+1}= {u}")

```
