```python
import numpy as np 
import pandas as pd 
```

```python
data= pd.read_csv("R_text.csv")
data.shape
print(data)
```

```python
A_Matrix= np.array(data.loc[:,'Ln(k)'])
A_matrix= A_Matrix.reshape(-1,1)
print(f' the A= \n {A_matrix}')
print(A_matrix.shape)
```

```python
B_Matrix= np.array(data.loc[:, ['ln(T)','1/T']])
ones_column = np.ones((A_matrix.shape[0], 1))
B_matrix= np.hstack((ones_column, B_Matrix))
print(f"B: \n {B_matrix}")
B_T= B_matrix.T
B_T_B= np.matmul(B_matrix.T,B_matrix)
B_T_B_inv= np.linalg.inv(B_T_B)
B_T_A= np.matmul(B_T, A_matrix)
try:
    X_matrix= np.matmul(B_T_B_inv,B_T_A)
    print(f"\n X is \n {X_matrix}")
except np.linalg.LinAlgError:
    print("Matrix is singulard cannot an be inverted.")
```

```python
# B * X = A 
try:
    X= np.linalg.lstsq(B_matrix, A_matrix, rcond=None)
    print(X)
except np.linalg.LinAlgError as e:
    # Handles singular or non-square matrices
    print("Error:", e)
```

```python
def solve_sliding_3rows(B,a):
    solution = []
    for i in range(B.shape[0]-2):
        A= B[i:i+3,:]
        re_a= a[i: i+3,]   
        x= np.linalg.solve(A, re_a)
        solution.append(x)
    return np.array(solution)

X_matrix = solve_sliding_3rows(B_matrix, A_matrix)
X_matrix
```

```python
def m_values(matrix):
    matrix=[]
    for j in range(len(X_matrix)):
        for i in range(len(X_matrix[j])):
            if (i == 1):
                matrix.append(X_matrix[j][i])
                print(f'x_matrix[{j},{i}]= {X_matrix[j][i]}')
            else:
                print("")
    return np.array(matrix)
m= m_values(X_matrix)
print(m.shape)
```
